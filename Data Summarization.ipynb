# Checking null values and filling 

```python
# to check, whether the table has null values, we will use, -- isnull()
data.isnull().sum()
```

```python
# Nullpercentage - it wil diplays the mean values of column.
data.isnull().mean()*100
```

```python
# value_counts() - is used to get the counted value from given df.
data.Survived.value_counts()
```

```python
# filling in null values, using "fillna"..  inplace=True - used to set data as permenant. otherwise, it wil change within next attempt.
data.Survived.fillna(0,inplace=True)
```

```python
data.isnull().sum()
```

```python
## now we are changing all null value existed colums . using fillna.
```

```python
data.Age.value_counts()
```

```python
data.Age.fillna(24.0,inplace=True)
```

```python
data.isnull().sum()
```

```python
data.Embarked.value_counts()
```

```python
data.Embarked.fillna('s',inplace=True)
```

```python
data.isnull().sum()
```

## Groupby


```python
# A groupby operation involves some combination of splitting the object, applying a function, and combining the results. This can be used to group large amounts of data and compute operations on these groups.
```

```python
Age_survived=data.groupby(['Age'])["Survived"].sum()
```

```python
Age_survived
```

```python
gender_survived=data.groupby(['Sex'])["Survived"].sum()
```

```python
gender_survived
```
